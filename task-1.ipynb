```python
import pandas as pd

df = pd.read_csv("/kaggle/input/datasets/amar5693/student-performance-dataset/ultimate_student_productivity_dataset_5000.csv")

filtered_data = df[df["study_hours"] > 5]

selected_data = df[["student_id", "study_hours", "exam_score"]]

subset_data = df.iloc[10:21, 0:5]

print("Filtered Rows:")
print(filtered_data.head())

print("\nSelected Columns:")
print(selected_data.head())

print("\nSubset Data:")
print(subset_data)
```
